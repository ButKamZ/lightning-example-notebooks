## <hr> Setup

```python
import os
from lightning import Lightning

from numpy import random, asarray, std
from sklearn import datasets
from scipy.ndimage.filters import gaussian_filter
```

```python
# replace with your own host and credentials, e.g. http://localhost:3000 or http://my-lightning-server.herokuapp.com
host = 'http://lightning-docs.herokuapp.com'
host = 'http://localhost:3000'
auth = (os.getenv('LIGHTNING_USERNAME'), os.getenv('LIGHTNING_PASSWORD'))
```

```python
lgn = Lightning(ipython=True, host=host, auth=auth)
lgn.create_session('scatterline-ipython');
```

## <hr> Spatial time series

When we have spatial and temporal data, with a time series for each of several spatial coordinates, scatter line plots are a great way to visualize both.
<br>
Scatter line plots take all the same stylings as scatter plots. Here we set the size based on the standard deviation.
<br>
Lightning is smart about fetching only the time series you are hovering over, to avoid loading all lines into the page.

```python
x = random.randn(20)
y = random.randn(20)
ts = gaussian_filter(random.rand(20,100), [0, 3])
s = std(ts, axis=1)*200
l = (random.rand(20)*5).astype('int')
lgn.scatterline(x, y, ts, label=l, size=s)
```

## <hr> Visualizing clusters

For multi-dimensional data, scatter plots only show two dimensions.
<br>
Scatter line plots are one (of many!) ways to look at more dimensions (as lines), while browsing the 2D representation.

```python
d, l = datasets.make_blobs(n_features=5, n_samples=200, centers=5, cluster_std=2.0, random_state=100)
x = d[:, 0]
y = d[:, 1]
lgn.scatterline(x, y, d, label=l, alpha=0.8, size=12)
```

## <hr> The iris data set

Here we use the same approach to examine the iris data set.

```python
iris = datasets.load_iris()
x = iris.data[:, 0]
y = iris.data[:, 1]
l = iris.target
lgn.scatterline(x, y, iris.data, label=l, alpha=0.8, size=12)
```
