## <hr> Setup

```python
import os
from lightning import Lightning

from numpy import random, asarray, amin, concatenate, column_stack 
from seaborn import color_palette
from sklearn import datasets
```

```python
# replace with your own host and credentials, e.g. http://localhost:3000 or http://my-lightning-server.herokuapp.com
host = 'http://lightning-docs.herokuapp.com'
auth = (os.getenv('LIGHTNING_USERNAME'), os.getenv('LIGHTNING_PASSWORD'))
```

```python
lgn = Lightning(ipython=True, host=host, auth=auth)
lgn.create_session('scatter3-ipython');
```

## <hr> Random points with default styling

```python
n = 100
x = random.rand(n)*100
y = random.rand(n)*100
z = random.rand(n)*100
lgn.scatter3(x,y,z)
```

## <hr> Random big red points

```python
n = 100
x = random.rand(n)*100
y = random.rand(n)*100
z = random.rand(n)*100
c = [240,117,145]
lgn.scatter3(x,y,z,size=10,color=c)
```

## <hr> Random points with all styling options

```python
n = 100
x = random.rand(n)*100
y = random.rand(n)*100
z = random.rand(n)*100
c = [asarray(color_palette('Blues', 100)[random.choice(range(100))])*255 for i in range(n)]
s = random.rand(n)*10+2
lgn.scatter3(x, y, z, color=c, size=s)
```

## <hr> Fun with colors

```python
n = 500
x = random.rand(n)*255
y = random.rand(n)*255
z = random.rand(n)*255
c = column_stack((x,y,z))
lgn.scatter3(x,y,z,color=c, size=5)
```
